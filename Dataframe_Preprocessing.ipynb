```python
#Import dependencies
import pandas as pd
```

```python
#Import and save each CSV file from resources

vaers_data_load = "Resources/2021VAERSDATA.csv"
vaers_symptoms_load= "Resources/2021VAERSSYMPTOMS.csv"
vaers_vax_load= "Resources/2021VAERSVAX.csv"
```

```python
# Create Dataframes
vaers_data_df = pd.read_csv(vaers_data_load,encoding='cp1252')
vaers_data_df.head()
```

```python
# Drop the CAGE_MO, RPT_DATE columns 
vaers_data_df.drop(columns = ["CAGE_MO", "RPT_DATE"], axis = 1)
```

```python
# Create Dataframes
vaers_symptoms_df = pd.read_csv(vaers_symptoms_load,encoding='cp1252')
vaers_symptoms_df.head()
```

```python
# Create Dataframes
vaers_vax_df = pd.read_csv(vaers_vax_load,encoding='cp1252')
vaers_vax_df.head()
```

```python
#Merge the dataframe into one on the VAERS_ID column
vaers_merge = pd.merge(vaers_data_df, vaers_symptoms_df, on=["VAERS_ID", "VAERS_ID",])
vaers_merge
```

```python
#Merge the dataframe into one on the VAERS_ID column
vaers_complete = pd.merge(vaers_merge, vaers_vax_df, on=["VAERS_ID", "VAERS_ID",])
vaers_complete
```

```python
#Export the merged csv

vaers_complete.to_csv('VAERSDATA.csv')
```
